儘管機器學習在1990年代才開始蓬勃發展，但它已迅速成為人工智能中最受歡迎合最成功的子領域，這一趨勢是由更快的硬件合更大的數據集推動的。

要定義深度學習並了解深度學習與其他機器學習方法之間的區別，首先我們需要了解機器學習算法的作用。<br>
機器學習會發現執行數據處理任務的規則，並給出預期的示例。因此要進行機器學習，我們需要三件事:<br>
1.輸入數據,如果任務是語音識別,這些數據可能是人們說話的聲音文件。如果任務是圖像識別,他們可能是圖片。<br>
2.預期輸入示例(標籤):在語音識別任務中,這些可能是人為生成的聲音文件副本。在圖像任務中，預期輸出可能是諸如"狗"、"貓"等標籤。<br>
3.一種衡量算法是否做得好的方法<br>







機器學習模型將其輸入數據轉換為有意義的輸出，這是一種從已知輸入和輸出示例中"學習"的過程。
因此，機器學習和深度學習的核心問題是有意義地轉換數據。

image.png

如果我們想開發一種算法，可以通過獲取一個點的座標，並輸出該點可能是黑色或白色。<br>
在這種情況下，<br>
1.輸出是點的座標<br>
2.預期的輸出是點的顏色<br>
3.衡量算法是否做得好的一種方法可以是，例如:正確分類的點的百分比。<br>
我們需要的是一種新的數據表示，可以清晰地將白點與黑點分開。<br>

image.png

深度學習如此迅速發展的主要原因是它為許多問題提供了更好的性能。<br>
但這不是唯一的原因，深度學習還使解決問題變得更加容易,因為它完全自動化了過去機器學習工作流程中最關鍵的步驟:"特徵工程"

以前的機器學習(淺層學習)只涉及將輸入數據轉換為一個或兩個連續的表示空間，通常只有簡單的轉換SVM或決策樹。但無法解決複雜的問題，因此，人類必須竭盡全力使初始輸入數據更易於通過這些方法進行處理:他們必須手動為其數據設計良好的表示層。這稱為特徵工程。<br>
深度學習完全自動化了這一步:通過深度學習，你可以一次學習所有特徵，而不必自己設計他們。這大大簡化了機器學習工作流程。<br>
